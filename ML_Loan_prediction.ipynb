#### Feature Scaling:

```python
loan_data.head(4)
```

```python
cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']
```

```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
X[cols] = st.fit_transform(X[cols])
```

```python
X.head(5)
```

#### Splitting The Dataset Into Training Set - Test Set & Applying K-Fold Cross Validation:

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
model_df={}
def model_val(model,X,y):
    X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size =0.20,
                                                    random_state =42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"{model} accuracy is {accuracy_score(y_test, y_pred)}")

    #5-fold Cross-Validation:
    score = cross_val_score(model, X, y, cv = 5)
    print(f"{model} Avg cross val score is {np.round(np.mean(score),3):.3f}")
    model_df[model]= round(np.mean(score)*100,2)

    # print(f"{model} Avg cross val score is {np.round(np.mean(score), 3):.3f}")
```

#### Logistic Regression:

```python
loan_data.head(4)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model_val(model,X,y)
```

```python
model_df
```

#### Support Vector Classifier (SVC):

```python
from sklearn import svm
model = svm.SVC()
model_val(model,X,y)
```

#### Decision Tree Classifier:

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model_val(model,X,y)
```

```python
model_df
```

#### Random Forest Classifier:

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model_val(model,X,y)
```

#### Gradient Boosting Classifier:

```python
from sklearn.ensemble import GradientBoostingClassifier
model =GradientBoostingClassifier()
model_val(model,X,y)
```

#### Result: Logistic Regression is the best model with 80.48 %

```python
model_df
```

#### Performing 'Hyperparameter' Tuning (for better accuracy):

```python
from sklearn.model_selection import RandomizedSearchCV
```

#### 1. Tune 'Hyper Parameter' for Logistic Regression:

```python
log_reg_grid = {"C":np.logspace(-4,4,20),
             "solver":['liblinear']}
```

```python
rs_log_reg = RandomizedSearchCV(LogisticRegression(),
                   param_distributions=log_reg_grid,
                  n_iter=20,cv=5,verbose=True)
```

```python
rs_log_reg.fit(X,y)
```

```python
rs_log_reg.best_score_
```

```python
rs_log_reg.best_params_
```

#### 2.'Hyper Parameter' tuning for SVC:

```python
svc_grid = {'C':[0.25,0.50,0.75,1],"kernel":["linear"]}
```

```python
rs_svc = RandomizedSearchCV(svm.SVC(), param_distributions =svc_grid, cv =5,
                   n_iter = 20, verbose = True)
```

```python
rs_svc.fit(X,y)
```

```python
rs_svc.best_score_
```

#### 3. 'Hyper Parameter' tuning for Random Forest Classifier:

```python
RandomForestClassifier()
```

```python
rf_grid = {'n_estimators':np.arange(10,1000,10),
           'max_features':['auto','sqrt'],
           'max_depth':[None,3,5,10,20,30],
           'min_samples_split':[2,5,20,50,100],
           'min_samples_leaf':[1,2,5,10]
 }
```

```python
rs_rf = RandomizedSearchCV(RandomForestClassifier(), param_distributions = rf_grid,
                   cv=5, n_iter=20, verbose =True)
```

```python
rs_rf.fit(X,y)
```

```python
rs_rf.best_params_
```

```python
rs_rf.best_score_
```
